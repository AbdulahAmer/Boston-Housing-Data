## Residuals: the difference between target and predicted values 
###  $$ r=y-\hat{y} $$ 

  Used to check if assumptions hold and model is valid, here we assume that a linear fit would work. 
   
   Residuals should be random, no clear pattern, if there is there is information that is important to make predictions in the residuals and we missed something. 
   
   For example you can get quadratic or some other fit, you can get an outlier, you can get weird linear or horizontal clustering, or even linear scaling of residuals, ie they get larger as the predicted value gets larger. 
   
   Ideal Residual graphs ( residuals vs $\hat{y} _{predict}$) are cloud likes they should be normally distrubuted. That means $Mean = 0 $ and $Skew = 0$. 

## Residual Plots 

```python
# Modified model, transformed using log prices & 
# simplified (dropping INDUS and AGE)
prices=np.log(data['PRICE']) # use log prices
features=data.drop(['PRICE', 'INDUS', 'AGE'], axis=1)

x_train, x_test, y_train, y_test = train_test_split(
                                    features, prices, 
                                    test_size=0.2, 
                                    random_state=10) 
#use statsmodel
x_include_constant = sm.add_constant(x_train)
model = sm.OLS(y_train, x_include_constant) 
results=model.fit()

#Residuals 
# residuals = y_train -results.fittedvalues
# residuals.describe()

#results object has resid attribute 
# results.resid

#Graph of Actual vs Predicted Prices 
corr = round(y_train.corr(results.fittedvalues), 3)
plt.scatter(x=y_train, y=results.fittedvalues, 
            c='navy', alpha=0.3)
plt.plot(y_train, y_train, color='cyan')



plt.xlabel('Actual log prices $ y_i$', fontsize=14)
plt.ylabel('Predicted log prices $ \hat y_i$', fontsize=14)
plt.title(f'Actual vs Predicted log prices: $y _i$  vs  $ \hat y_i$  corr: {corr}')

plt.show()

plt.scatter(x=np.e**y_train, 
            y=np.e**results.fittedvalues, c='darkred', 
            alpha=0.3)
plt.plot(np.e**y_train, np.e**y_train, color='magenta')



plt.xlabel('Actual prices $ y_i$', fontsize=14)
plt.ylabel('Predicted prices $ \hat y_i$', fontsize=14)
plt.title(f'Actual vs Predicted log prices: $y _i$  vs  $ \hat y_i$  corr: {corr}')

plt.show()

# Residuals vs predicted values 
plt.scatter(x=results.fittedvalues, y=results.resid, 
            cmap='viridis', alpha=0.6)


plt.xlabel('Fitted Values ', fontsize=14)
plt.ylabel('Residuals$', fontsize=14)
plt.title('Residuals vs Fitted Values')

plt.show()

#Mean Squared Error & R^2
reduced_log_mse=round(results.mse_resid,3)
reduced_log_rsquared=round(results.rsquared,3)
```

```python
#Distribution of Residuals (log prices)  
#checking for normality
mean =round(results.resid.mean(),3)
skew= round(results.resid.skew(),3)
print('mean: ', mean )
print('skew: ', skew )

sns.distplot(results.resid)
plt.title(f'Log price model: Residuals Skew({skew}), Mean({mean})')
plt.show()
# shows us this data is pretty normal 
```

```python
pd.DataFrame({'R-Squared' : [reduced_log_rsquared, 
                             full_normal_rsquared, 
                            omitted_var_rsquared], 
             'MSE': [reduced_log_mse, 
                     full_normal_mse, 
                     omitted_var_mse], 
             'RMSE': np.sqrt([reduced_log_mse, 
                     full_normal_mse, 
                     omitted_var_mse])}, 
            index=['Reduced Log Model',
                   'Full Normal Price Model', 
                   'Omitted Var Model'])

# R squared is realtive mesaure of fit 
# MSE is absolute 
# scale of MSE in omittied and reduced log is log not 
# actual MSE in full normal 

```

```python
# 68% of observations in a normal dist are 
# between 1 and -1 std from the mean 
# 95% between 2 and -2 std from the mean
# use dist from residuals that we got 
# square root mse RMSE is used to extend bounds 
# of a dist 

print( ' 1 std in log prices is:  ', np.sqrt(reduced_log_mse))
print( ' 2 std in log prices is:  ', 2*np.sqrt(reduced_log_mse))

upper_bound = np.log(30) + 2*np.sqrt(reduced_log_mse)
print('The upper bound in Log prices for a 95% prediciton interval is: ', upper_bound)
print('The upper bound in normal prices is: ', np.e**upper_bound)

lower_bound = np.log(30) - 2*np.sqrt(reduced_log_mse)
print('The lower bound in Log prices for a 95% prediciton interval is: ', lower_bound)
print('The lower bound in normal prices is: ', np.e**lower_bound)
```
